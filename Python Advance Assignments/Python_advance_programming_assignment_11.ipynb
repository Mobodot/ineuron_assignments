```python
"""Question 1:
Create a function that takes a list and returns a new 
list containing only prime numbers.

Examples

filter_primes([7, 9, 3, 9, 10, 11, 27]) ➞ [7, 3, 11]

filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]) ➞ [10007, 1009]

filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097]) ➞ [1009, 3, 61, 1087, 1091, 1093, 1097]
"""

def is_prime(number):
    """
    
    Returns True if a number is a prime number otherwise
    False.
    
    """
    if number <= 1:
        return False
    for digit in range(2, number):
        if number % digit == 0:
            return False
    return True

        
def filter_primes(lst):
    """
    
    Returns a list prime number from a given list.
    
    """
    prime_no = []
    for value in lst:
        try:
            if is_prime(value):
                prime_no.append(value)
        except Exception as e:
            return "Invalid value present", e
    return prime_no



lst1 = [7, 9, 3, 9, 10, 11, 27]
lst2 = [10007, 1009, 1007, 27, 147, 77, 1001, 70]
lst3 = [1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097]

print("----- A Program to Filter the Prime Numbers in a List -----")
print()
print(f"Prime numbers in {lst1} = {filter_primes(lst1)}")
print(f"Prime numbers in {lst2} = {filter_primes(lst2)}")
print(f"Prime numbers in {lst3} = {filter_primes(lst3)}")
```

```python
"""Question 2:
Once a water balloon pops, is soaks the area around it. 
The ground gets drier the further away you travel from 
the balloon.

The effect of a water balloon popping can be modeled 
using a list. Create a function that takes a list which 
takes the pre-pop state and returns the state after the 
balloon is popped. The pre-pop state will contain at most
a single balloon, whose size is represented by the only 
non-zero element.

Examples

pop([0, 0, 0, 0, 4, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4, 3, 2, 1, 0]

pop([0, 0, 0, 3, 0, 0, 0]) ➞ [0, 1, 2, 3, 2, 1, 0]

pop([0, 0, 2, 0, 0]) ➞ [0, 1, 2, 1, 0]

pop([0]) ➞ [0]
"""

def pop(lst):
    """
    
    Takes a list with only 2 distinct values and returns
    a new list which show the increment and decrement in
    the list.
    
    If the number of distinct values is greater than 2
    returns [0].
    
    Examples
    
    pop([0, 0, 0, 3, 0, 0, 0]) ➞ [0, 1, 2, 3, 2, 1, 0]

    pop([0, 0, 2, 0, 0]) ➞ [0, 1, 2, 1, 0]

    pop([0]) ➞ [0]
    
    """
    post_pop = []
    unique_values = list(set(lst))
    counter = 0
    
    if len(unique_values) == 2:
        for value in range(len(lst)):
            if value <= lst[unique_values[1]]:
                post_pop.append(value)
            else:
                counter += 2
                post_pop.append(value - counter)
        return post_pop
    return [0]

lst1 = [0, 0, 0, 0, 4, 0, 0, 0, 0]
lst2 = [0, 0, 0, 3, 0, 0, 0]
lst3 = [0, 0, 2, 0, 0]
lst4 = [0]

print("---- A program to Show the Post Effect of a Single Balloon ----")
print()
print(f"Post pop effect of {lst1} = {pop(lst1)}")
print(f"Post pop effect of {lst2} = {pop(lst2)}")
print(f"Post pop effect of {lst3} = {pop(lst3)}")
print(f"Post pop effect of {lst4} = {pop(lst4)}")
  
```

```python
"""Question 3:
"Loves me, loves me not" is a traditional game in which a
person plucks off all the petals of a flower one by one, 
saying the phrase "Loves me" and "Loves me not" when 
determining whether the one that they love, loves them 
back.

Given a number of petals, return a string which repeats 
the phrases "Loves me" and "Loves me not" for every 
alternating petal, and return the last phrase in all 
caps. Remember to put a comma and space between phrases.

Examples

loves_me(3) ➞ "Loves me, Loves me not, LOVES ME"

loves_me(6) ➞ "Loves me, Loves me not, Loves me, Loves me not, Loves me, LOVES ME NOT"

loves_me(1) ➞ "LOVES ME"
"""

def phrase(number):
    """
    
    Returns loves me if a number is odd, else if a number
    is even return love me not.
    
    """
    if number % 2 != 0:
        return "loves me"
    return "loves me not"

def loves_me(number):
    """
    
    Returns a string containing the phrase loves me or 
    loves me not and capitalizes the last phrase in the
    string.
    
    """
    sentence = []
    if number > 0:
        for value in range(1, number+1):
            if value == number:
                sentence.append(phrase(value).upper())
            else:
                sentence.append(phrase(value))
        return ", ".join(sentence)
    return "Value must be greater than 0"
    

print("----- A Program to Determine Whether a Person Loves you or not ----")
print()
while True:

    try:
        user_input = int(input("Enter number : "))
    except Exception as e:
        print("Value must be an integer")
    else:
        print(f"Result when value is {user_input} = {loves_me(user_input)}")
        break
```

```python
"""Question 4:
Write a function that sorts each string in a list by the 
letter in alphabetic ascending order (a-z).

Examples

sort_by_letter(["932c", "832u32", "2344b"])
➞ ["2344b", "932c", "832u32"]

sort_by_letter(["99a", "78b", "c2345", "11d"])
➞ ["99a", "78b", "c2345", "11d"]

sort_by_letter(["572z", "5y5", "304q2"])
➞ ["304q2", "5y5", "572z"]

sort_by_letter([])
➞ []
"""

def letters_list(lst):
    """
    
    Extracts the letters from a list of string and returns
    an ordered list of letters.
    
    """
    letters = []
    for string in lst:
        for value in string:
            if value.isalpha():
                letters.append(value)
                break
    return sorted(letters)

letters_list(["572z", "5y5", "304q2"])
    
    
def sort_by_letter(lst):
    """
    
    Returns a list of sorted strings based on the letters
    in alphabetical ascending order.
    
    Examples

    sort_by_letter(["932c", "832u32", "2344b"])
    ➞ ["2344b", "932c", "832u32"]

    sort_by_letter(["99a", "78b", "c2345", "11d"])
    ➞ ["99a", "78b", "c2345", "11d"]
    
    """
    ordered_list = []
    for letter in range(len(letters_list(lst))):     # loop through list of ordered letters
        for string in range(len(lst)):     # loop through list of strings
            if letters_list(lst)[letter] in lst[string]:     # check if ordered letter is in list of strings
                ordered_list.append(lst[string])
                break
    return ordered_list

print("----- A Program Which sorts the Strings in a list \
Based on Alphabetical Ascending Order (a-z) -----")

lst1 = ["932c", "832u32", "2344b"]
lst2 = ["99a", "78b", "c2345", "11d"]
lst3 = ["572z", "5y5", "304q2"]
lst4 = []

print()
print(f"Sorted {lst1} = {sort_by_letter(lst1)}")
print(f"Sorted {lst2} = {sort_by_letter(lst2)}")
print(f"Sorted {lst3} = {sort_by_letter(lst3)}")
print(f"Sorted {lst4} = {sort_by_letter(lst4)}")
```

```python
"""Question 5:
There are three cups on a table, at positions A, B, and 
C. At the start, there is a ball hidden under the cup at 
position B.
However, I perform several swaps on the cups, which is 
notated as two letters. For example, if I swap the cups 
at positions A and B, I could notate this as AB or BA.

Create a function that returns the letter position that 
the ball is at, once I finish swapping the cups. The 
swaps will be given to you as a list.

Example

cup_swapping(["AB", "CA", "AB"]) ➞ "C"

# Ball begins at position B.
# Cups A and B swap, so the ball is at position A.
# Cups C and A swap, so the ball is at position C.
# Cups A and B swap, but the ball is at position C, so it
doesn't move.

"""

def cup_swapping(lst):
    """
    
    Returns a letter which shows the position of a ball
    after swapping cups.
    
    """
    try:
        for string in range(len(lst)-1):
            # if first letter in string = second letter next string, it has been swapped
            if lst[string][0] == lst[string+1][1]:
                continue
            else:     # it hasn't been swapped
                return lst[string][0]
        return lst[string+1][0]
    except TypeError as e:
        return e
    except Exception as e:
        return e
        

lst1 = ["AB", "CA", "AB"]
lst2 = ["AB", "CA", "BC"]
print(f"Postion of ball after swapping in {lst1} = {cup_swapping(lst1)}")
print(f"Postion of ball after swapping in {lst2} = {cup_swapping(lst2)}")
```
