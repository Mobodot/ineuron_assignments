```python
"""Question 1: 
Write a function that stutters a word as if someone is 
struggling to read it. The first two letters are 
repeated twice with an ellipsis ... and space after each,
and then the word is pronounced with a question mark ?.

Examples
stutter("incredible") ➞ "in... in... incredible?"

stutter("enthusiastic") ➞ "en... en... enthusiastic?"

stutter("outstanding") ➞ "ou... ou... outstanding?"

Hint :- Assume all input is in lower case and at least 
two characters long.
"""

def stutter(string):
    """
    
    Returns a string which has been stuttered.
    
    E.g:
    stutter("incredible") ➞ "in... in... incredible?"

    stutter("enthusiastic") ➞ "en... en... enthusiastic?"

    stutter("outstanding") ➞ "ou... ou... outstanding?"
    
    """
    string_list = []
    first_two_words_with_ellipsis = string[:2] + "... "
    counter = 0
    
    while counter < 3:
        # puts the whole word into list
        if counter == 2:
            string_list.append(string + "?")
        else:
            # puts stutter into list
            string_list.append(first_two_words_with_ellipsis)
        counter += 1
    return "".join(string_list)


print("A Program to Stutter or Repeat a String")
print()

while True:
    try:
        print()
        user_input = input("Enter a string/word: ")
        if len(user_input) < 3:     # len of word must be greater than 2
            print("Invalid input! String must contain more than two letters.")
            continue
    except Exception as e:
        pass
    else:
        print(f"Stuttered/Repeated String ===> {stutter(user_input)}")
        break
```

```python
"""Question 2
Create a function that takes an angle in radians and 
returns the corresponding angle in degrees rounded to 
one decimal place.

Examples
radians_to_degrees(1) ➞ 57.3

radians_to_degrees(20) ➞ 1145.9

radians_to_degrees(50) ➞ 2864.8"""

def radians_to_degrees(num):
    """
    
    Returns a value after converting radian to degrees.
    
    """
    return round((int(num)*180)/3.142, 1)

print("A Program to Convert Radian to Degrees")
print()

while True:
    try:
        print()
        user_input = input("Enter a digit: ")
        if not user_input.isdigit():     # ensures user enters a digit
            print("Invalid input! Input must be numeric.")
            continue
    except Exception as e:
        pass
    else:
        print(f"{user_input}radian to degree = {radians_to_degrees(user_input)}degrees")
        break
```

```python
"""Question 3:
In this challenge, establish if a given integer num is a
Curzon number. If 1 plus 2 elevated to num is exactly 
divisible by 1 plus 2 multiplied by num, then num is a 
Curzon number.
Given a non-negative integer num, implement a function 
that returns True if num is a Curzon number, or False 
otherwise.

Examples
is_curzon(5) ➞ True
# 2 ** 5 + 1 = 33
# 2 * 5 + 1 = 11
# 33 is a multiple of 11

is_curzon(10) ➞ False
# 2 ** 10 + 1 = 1025
# 2 * 10 + 1 = 21
# 1025 is not a multiple of 21

is_curzon(14) ➞ True
# 2 ** 14 + 1 = 16385
# 2 * 14 + 1 = 29
# 16385 is a multiple of 29
"""

def is_curzon(num):
    """
    
    Returns True if a number is curzon.
    
    
    """
    numerator = (2 ** int(num)) + 1
    denominator = (2 * int(num)) + 1
    if numerator % denominator  == 0:
        return True
    else:
        return False
    

    
print("A Program to Check if a Number is Curzon.")
print()

while True:
    try:
        print()
        user_input = input("Enter a digit: ")
        if not user_input.isdigit():     # ensures user enters a digit
            print("Invalid input! Input must be numeric.")
            continue
    except Exception as e:
        pass
    else:
        print(f"{user_input} number is Curzon = {is_curzon(user_input)}")
        break
```

```python
"""Question 4:
Given the side length x find the area of a hexagon.

Examples
area_of_hexagon(1) ➞ 2.6

area_of_hexagon(2) ➞ 10.4

area_of_hexagon(3) ➞ 23.4"""

def area_of_hexagon(side_length):
    
    from math import sqrt
    
    area = 3 * sqrt(3) * int(side_length)**2 / 2
    return round(area, 1)


print("A Program to Find the Area of a Hexagon.")
print()

while True:
    try:
        print()
        user_input = input("Enter the side_length: ")
        if not user_input.isdigit():     # ensures user enters a digit
            print("Invalid input! Input must be numeric.")
            continue
    except Exception as e:
        pass
    else:
        print(f"Area of Hexagon with length {user_input} = {area_of_hexagon(user_input)}")
        break
```

```python
"""Question 5
Create a function that returns a base-2 (binary) 
representation of a base-10 (decimal) string number. To 
convert is simple: ((2) means base-2 and (10) means 
base-10) 010101001(2) = 1 + 8 + 32 + 128.
Going from right to left, the value of the most right 
bit is 1, now from that every bit to the left will be x2 
the value, value of an 8 bit binary numbers are 
(256, 128, 64, 32, 16, 8, 4, 2, 1).

Examples
binary(1) ➞ "1"
# 1*1 = 1

binary(5) ➞ "101"
# 1*1 + 1*4 = 5

binary(10) ➞ "1010"
# 1*2 + 1*8 = 10
"""

def binary(num):
    """
    
    Returns the binary representation of a number in a str.
    
    E.g: 5 = "101"
    
    """
    quotient = None
    divisor = 2
    dividend = num
    remainders = []

    # produces the remainders
    while quotient != 0:
        quotient, remainder = divmod(dividend, divisor)
        remainders.append(remainder)

        dividend = quotient


    # ordering the remainders from bottom to top
    ordered_remainders = []
    for value in range(len(remainders)):
        ordered_remainders.append(str(remainders.pop()))    # changed each value to str the can be joined to a single value

    return "".join(ordered_remainders)


print("A Program to Produce the Binary Rep of a Number.")

while True:
    try:
        print()
        user_input = int(input("Enter a number: "))
    except Exception as e:
        pass
    else:
        print(f"Binary rep of {user_input}  = {binary(user_input)}")
        break
```
