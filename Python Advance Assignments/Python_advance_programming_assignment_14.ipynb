```python
"""Question 1:
Given a list of numbers, create a function that removes 
25% from every number in the list except the smallest 
number, and adds the total amount removed to the smallest
number.

Examples

show_the_love([4, 1, 4]) ➞ [3, 3, 3]

show_the_love([16, 10, 8]) ➞ [12, 7.5, 14.5]

show_the_love([2, 100]) ➞ [27, 75]
"""

def deduction(lst):
    """
    
    Returns a total of 25% deduction from all the numbers
    in a list except the smallest number.
    
    """
    smallest = min(lst)
    deduction = 0
    for value in lst:
        if value != min(lst):
            deduction += (0.25 * value)
    return deduction


def show_the_love(lst):
    """
    
    Returns a list which deducts 25% from each number
    and adds the total of all deductions to the smallest
    number.
    
    Examples

    show_the_love([4, 1, 4]) ➞ [3, 3, 3]

    show_the_love([16, 10, 8]) ➞ [12, 7.5, 14.5]
    
    """
    values = []
    try:
        for value in lst:
            if value == min(lst):
                values.append(value + deduction(lst))
            else:
                values.append(value - (0.25 * value))
        return values
    except TypeError as e:
        return f"Value must be int not {type(value)}"
    except Exception as e:
        return e


print("----- A Program which removes 25% from all the numbers except\
 the smallest and adds the total amount removed to the smallest Number ----")
print()

lst1 = [4, 1, 4]
lst2 = [16, 10, 8]
lst3 = [2, 100]

print(f"Deductions from {lst1} = {show_the_love(lst1)}")
print(f"Deductions from {lst2} = {show_the_love(lst2)}")
print(f"Deductions from {lst3} = {show_the_love(lst3)}")
```

```python
"""Question 2:
Create a function that takes in two words as input and 
returns a list of three elements, in the following order:

   1.Shared letters between two words.
   2.Letters unique to word 1.
   3.Letters unique to word 2.

Each element should have unique letters, and have each 
letter be alphabetically sorted.

Examples

letters("sharp", "soap") ➞ ["aps", "hr", "o"]

letters("board", "bored") ➞ ["bdor", "a", "e"]

letters("happiness", "envelope") ➞ ["enp", "ahis", "lov"]

letters("kerfuffle", "fluffy") ➞ ["flu", "ekr", "y"]
# Even with multiple matching letters (e.g. 3 f's), 
there should 
# only exist a single "f" in your first element.

letters("match", "ham") ➞ ["ahm", "ct", ""]
# "ham" does not contain any letters that are not found already 
# in "match".
"""

def letters(string_a, string_b):
    """
    
    Returns a list containing:
    1.Shared letters between two words.
    2.Letters unique to word 1.
    3.Letters unique to word 2.
    
    NB: Elements have unique letters, and each letter 
    is alphabetically sorted.
    
    Examples

    letters("sharp", "soap") ➞ ["aps", "hr", "o"]

    letters("board", "bored") ➞ ["bdor", "a", "e"]
    
    """
    similar_letters = []
    unique_string_a = []
    unique_string_b = []
    
    # check if similar letters exist in both strings and ensure 
    # duplicate letters are not stored.
    try:
        for letter in string_a:
            if letter in string_b and letter not in similar_letters:
                similar_letters.append(letter)
            elif letter not in string_b and letter not in unique_string_a:
                unique_string_a.append(letter)

        # check for unique elements in string_b and ensure duplicate
        # letters are not stored.
        for letter in string_b:
            if letter not in string_a and letter not in unique_string_b:
                unique_string_b.append(letter)

        return ["".join(sorted(similar_letters)), 
                "".join(sorted(unique_string_a)), 
                "".join(sorted(unique_string_b))]
    except TypeError as e:
        return e
    except Exception as e:
        return e


string_a1, string_a2 = "sharp", "soap"
string_b1, string_b2 = "board", "bored"
string_c1, string_c2 = "happiness", "envelope"
string_d1, string_d2 = "kerfuffle", "fluffy"
string_e1, string_e2 = "match", "ham"

print(f"Shared, unique letters in first word and unique letters\
in second word in {string_a1, string_a2} = {letters(string_a1, string_a2)}")
print()
print(f"Shared, unique letters in first word and unique letters\
in second word in {string_b1, string_b2} = {letters(string_b1, string_b2)}")
print()
print(f"Shared, unique letters in first word and unique letters\
in second word in {string_c1, string_c2} = {letters(string_c1, string_c2)}")
print()
print(f"Shared, unique letters in first word and unique letters\
in second word in {string_d1, string_d2} = {letters(string_d1, string_d2)}")
print()
print(f"Shared, unique letters in first word and unique letters\
in second word in {string_e1, string_e2} = {letters(string_e1, string_e2)}")
```

```python
"""Question 3:
Write a function that pairs the first number in an array 
with the last, the second number with the second to last,
etc.

Examples

pairs([1, 2, 3, 4, 5, 6, 7]) ➞ [[1, 7], [2, 6], [3, 5], [4, 4]]

pairs([1, 2, 3, 4, 5, 6]) ➞ [[1, 6], [2, 5], [3, 4]]

pairs([5, 9, 8, 1, 2]) ➞ [[5, 2], [9, 1], [8, 8]]

pairs([]) ➞ []
"""

def pairs(lst):
    """
    
    Returns a paired list which is a combination of the 
    first and last number, second and second to the last
    number of an array and so on.
    
    Examples

    pairs([1, 2, 3, 4, 5, 6, 7]) ➞ [[1, 7], [2, 6], [3, 5], [4, 4]]

    pairs([1, 2, 3, 4, 5, 6]) ➞ [[1, 6], [2, 5], [3, 4]]
    
    """
    value_pairs = []
    if len(lst) > 2:
        for value in range((len(lst)//2)+1):
            value_pairs.append([lst[value], lst[len(lst)-1-value]])
        return value_pairs
    return lst


print("----- A Program to pair the numbers in a list -----")
print()

lst1 = [1, 2, 3, 4, 5, 6, 7]
lst2 = [1, 2, 3, 4, 5, 6]
lst3 = [5, 9, 8, 1, 2]
lst4 = []

print(f"Pairs in {lst1} = {pairs(lst1)}")
print(f"Pairs in {lst2} = {pairs(lst2)}")
print(f"Pairs in {lst3} = {pairs(lst3)}")
print(f"Pairs in {lst4} = {pairs(lst4)}")
```

```python
"""Question 4:
Write a function that adds two numbers. The catch, 
however, is that the numbers will be strings.

Examples

add_str_nums("4", "5") ➞ "9"

add_str_nums("abcdefg", "3") ➞ "-1"

add_str_nums("1", "") ➞ "1"

add_str_nums("1874682736267235927359283579235789257", "32652983572985729") ➞ "1874682736267235927391936562808774986"
"""

def add_str_nums(num1, num2):
    """
    
    Returns the addition of two numbers which are strings.
    If either of the values is not a string return -1.
    
    """
    if type(num1) == str and type(num2) == str:
        if num1.isdigit() and num2.isdigit():
            return str(int(num1) + int(num2))
        elif num1 == "" or num2 == "":
            return num1 + num2
        elif num1.isalpha() or num2.isalpha():
            return "-1"
    else:
        return "Values must be given as strings!"
    
    
print("----- A Program to add Two Numeric Strings together -----")
print()

num1a, num1b = "4", "5"
num2a, num2b = "abcdefg", "3"
num3a, num3b = "1", ""
num4a, num4b = "1874682736267235927359283579235789257", "32652983572985729"

print(f"Addition of {num1a} and {num1b} = {add_str_nums(num1a, num1b)}")
print(f"Addition of {num2a} and {num2b} = {add_str_nums(num2a, num2b)}")
print(f"Addition of {num3a} and {num3b} = {add_str_nums(num3a, num3b)}")
print(f"Addition of {num4a} and {num4b} = {add_str_nums(num4a, num4b)}")
```

```python
"""Question 5:
lPaeesh le pemu mnxit ehess rtnisg! Oh, sorry, that was 
supposed to say: Please help me unmix these strings!

Somehow my strings have all become mixed up; every pair 
of characters has been swapped. Help me undo this so I 
can understand my strings again.

Examples

unmix("123456") ➞ "214365"

unmix("hTsii  s aimex dpus rtni.g") ➞ "This is a mixed up string."

unmix("badce") ➞ "abcde"
"""

def unmix(string):
    """
    
    Takes a mixed string and returns the correct meaning.
    
    """
    if type(string) == str:
        new_string = []
        for value in range(len(string)//2):
            new_string.append(string[value+1+value])
            new_string.append(string[value+value])

        return "".join(new_string)
    else:
        return "Value must be a string"


print("----- A Program to take a mixed string and Output the Correct Meaning -----")
print()

string_a, string_b, string_c = "123456", "hTsii  s aimex dpus rtni.g", "abcde"

print(f"Correct meaning of {string_a} = {unmix(string_a)}")
print(f"Correct meaning of {string_b} = {unmix(string_b)}")
print(f"Correct meaning of {string_c} = {unmix(string_c)}")

```
