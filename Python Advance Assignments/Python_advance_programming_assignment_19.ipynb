```python
"""Question 1:
Create a checker board generator, which takes as inputs
n and 2 elements to generate an n x n checkerboard with
those two elements as alternating squares.

Examples

checker_board(2, 7, 6) ➞ [
  [7, 6],
  [6, 7]
]

checker_board(3, "A", "B") ➞ [
  ["A", "B", "A"],
  ["B", "A", "B"],
  ["A", "B", "A"]
]

checker_board(4, "c", "d") ➞ [
  ["c", "d", "c", "d"],
  ["d", "c", "d", "c"],
  ["c", "d", "c", "d"],
  ["d", "c", "d", "c"]
]

checker_board(4, "c", "c") ➞ "invalid"
"""

def checker_board(n, value_a, value_b):
    """
    
    n: int
    value_a: numeric or string value
    value_b: numeric or string value
    
    Returns a checker board of n x n if value_a and value_b
    are distinct. Otherwise returns invalid
    
    Examples

    checker_board(2, 7, 6) ➞ [
      [7, 6],
      [6, 7]
    ]
    
    checker_board(3, "A", "B") ➞ [
      ["A", "B", "A"],
      ["B", "A", "B"],
      ["A", "B", "A"]
    ]

    """
    try:
        major_list = []
        if value_a != value_b:
            for row in range(n):
                minor_list = []
                if row % 2 == 0:     # insertion starts with first value for even rows
                    for col in range(n):
                        if col % 2 == 0:
                            minor_list.append(value_a)
                        else:
                            minor_list.append(value_b) 
                    major_list.append(minor_list)
                else:
                    for col in range(n):     # insertion starts with second value for odd rows
                        if col % 2 == 0:
                            minor_list.append(value_b)
                        else:
                            minor_list.append(value_a)
                    major_list.append(minor_list)
            return major_list
        return "invalid"
    
    except TypeError as e:
        return f"n value must be an int not {type(n)}"
    except Exception as e:
        return e
            
        
n_a, value_1a, value_1b = 2, 7, 6
n_b, value_2a, value_2b = 3, "A", "B"
n_c, value_3a, value_3b = 4, "c", "d"
n_d, value_4a, value_4b = 4, "c", "c"

print("-----A program to generate a checker board-----")
print()

print(f"Check board using values {n_a}, {value_1a}, {value_1b} = {checker_board(n_a, value_1a, value_1b)}")
print(f"Check board using values {n_b}, {value_2a}, {value_2b} = {checker_board(n_b, value_2a, value_2b)}")
print(f"Check board using values {n_c}, {value_3a}, {value_3b} = {checker_board(n_c, value_3a, value_3b)}")
print(f"Check board using values {n_d}, {value_4a}, {value_4b} = {checker_board(n_d, value_4a, value_4b)}")
```

```python
"""Question 2:
A string is an almost-palindrome if, by changing only 
one character, you can make it a palindrome. Create a 
function that returns True if a string is an 
almost-palindrome and False otherwise.

Examples

almost_palindrome("abcdcbg") ➞ True
# Transformed to "abcdcba" by changing "g" to "a".

almost_palindrome("abccia") ➞ True
# Transformed to "abccba" by changing "i" to "b".

almost_palindrome("abcdaaa") ➞ False
# Can't be transformed to a palindrome in exactly 1 turn.

almost_palindrome("1234312") ➞ False
"""

def almost_palindrome(string):
    """
    
    Returns True if a string is almost a palindrome,
    Otherwise False.
    
    A string is an almost-palindrome if, by changing only 
    one character, you can make it a palindrome.
    
    Examples

    almost_palindrome("abcdcbg") ➞ True
    # Transformed to "abcdcba" by changing "g" to "a".

    almost_palindrome("abccia") ➞ True
    # Transformed to "abccba" by changing "i" to "b".

    almost_palindrome("abcdaaa") ➞ False
    # Can't be transformed to a palindrome in exactly 1 turn.
    
    """
    try:
        reversed_str = string[::-1]
        for value in range(len(string)):
            if string[value] != reversed_str[value]:
                string = string.replace(reversed_str[value], string[value], 1)
                break

        if string == string[::-1]:
            return True
        return False
    
    except Exception as e:
        return e
        
str_a, str_b, str_c, str_d = "abcdcbg", "abccia", "abcdaaa", "1234312"

print("A Program to check if a string is almost palindrome or not")
print()

print(f"'{str_a}' is almost a palindrome = {almost_palindrome(str_a)}")
print(f"'{str_b}' is almost a palindrome = {almost_palindrome(str_b)}")
print(f"'{str_c}' is almost a palindrome = {almost_palindrome(str_c)}")
print(f"'{str_d}' is almost a palindrome = {almost_palindrome(str_d)}")
       
```

```python
"""Question 3:
Create a function that finds how many prime numbers 
there are, up to the given integer.

Examples

prime_numbers(10) ➞ 4
# 2, 3, 5 and 7

prime_numbers(20) ➞ 8
# 2, 3, 5, 7, 11, 13, 17 and 19

prime_numbers(30) ➞ 10
# 2, 3, 5, 7, 11, 13, 17, 19, 23 and 29
"""

def is_prime(num):
    """
    
    Return True is a number is prime, otherwise False.
    
    """
    if num > 1:
        for value in range(2, num):
            if num % value == 0:
                return False
        return True
    return False


def prime_numbers(num):
    """
    
    Returns the count of prime number in the range of
    a number.
    
    Examples

    prime_numbers(10) ➞ 4
    # 2, 3, 5 and 7

    prime_numbers(20) ➞ 8
    # 2, 3, 5, 7, 11, 13, 17 and 19
    
    """
    try:
        counter = 0
        for value in range(num):
            if is_prime(value):
                counter += 1

        return counter
    
    except TypeError as e:
        return f"value must be an int not {type(num)}"
    except Exception as e:
        return e

num1, num2, num3 = 10, 20, 30

print("A program to find how many prime numbers are there up to the given integer")
print()

print(f"Count of prime numbers in range {num1} = {prime_numbers(num1)}")
print(f"Count of prime numbers in range {num2} = {prime_numbers(num2)}")
print(f"Count of prime numbers in range {num3} = {prime_numbers(num3)}")
```

```python
"""Question 4:
If today was Monday, in two days, it would be Wednesday.

Create a function that takes in a list of days as input
and the number of days to increment by. Return a list of
days after n number of days has passed.

Examples

after_n_days(["Thursday", "Monday"], 4) ➞ ["Monday", "Friday"]

after_n_days(["Sunday", "Sunday", "Sunday"], 1) ➞ ["Monday", "Monday", "Monday"]

after_n_days(["Monday", "Tuesday", "Friday"], 1) ➞ ["Tuesday", "Wednesday", "Saturday"]
"""

def after_n_days(lst_days, n):
    """
    
    lst_days: list of days (list)
    n: increment (int)
    
    Returns a list of days which is an increment after
    n, days.
    
    Examples

    after_n_days(["Thursday", "Monday"], 4) ➞ ["Monday", "Friday"]

    after_n_days(["Sunday", "Sunday", "Sunday"], 1) ➞ ["Monday", "Monday", "Monday"]

    """
    days = []
    const_days = ["Sunday", "Monday", "Tuesday", "Wednesday",
           "Thursday", "Friday", "Saturday"]
    
    if type(lst_days) == list:
        try:
            for actual_day in lst_days:
                incrm = const_days.index(actual_day.title()) + n
                days.append(const_days[incrm%7])
            return days
        
        except Exception as e:
            return e
    return "lst_days must be a list!"


days_1, n_1 = ["Thursday", "Monday"], 4
days_2, n_2 = ["Sunday", "Sunday", "Sunday"], 1
days_3, n_3 = ["Monday", "Tuesday", "Friday"], 1

print("----- A program to display the day(s) after n days -----")
print()

print(f"Given {days_1} after {n_1} days = {after_n_days(days_1, n_1)}")
print(f"Given {days_2} after {n_2} days = {after_n_days(days_2, n_2)}")
print(f"Given {days_3} after {n_3} days = {after_n_days(days_3, n_3)}")
```
