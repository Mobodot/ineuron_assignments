```python
"""Question 1:
Rondo Form is a type of musical structure, in which 
there is a recurring theme/refrain, notated as A. Here 
are the rules for valid rondo forms:

   - Rondo forms always start and end with an A section.
   - In between the A sections, there should be 
   contrasting sections notated as B, then C, then D, 
   etc... No letter should be skipped.
   - There shouldn't be any repeats in the sequence 
   (such as ABBACCA).

Create a function which validates whether a given 
string is a valid Rondo Form.

Examples

valid_rondo("ABACADAEAFAGAHAIAJA") ➞ True

valid_rondo("ABA") ➞ True

valid_rondo("ABBACCA") ➞ False

valid_rondo("ACAC") ➞ False

valid_rondo("A") ➞ False
"""

def valid_rondo(string):
    """
    
    Returns True if a string is in a Rondo otherwise 
    False.
    
    The Rules for a valid rondo forms are:
    - Rondo forms always start and end with an A section.
    - In between the A sections, there should be 
    contrasting sections notated as B, then C, then D, 
    etc... No letter should be skipped.
    - There shouldn't be any repeats in the sequence 
    (such as ABBACCA)
    
    Examples

    valid_rondo("ABACADAEAFAGAHAIAJA") ➞ True

    valid_rondo("ABA") ➞ True

    valid_rondo("ABBACCA") ➞ False
    
    valid_rondo("A") ➞ False
    
    """
    try:
        if len(string) % 2 == 1 and len(string) > 1:
            for letter in range(0, len(string), 2):
                if string[letter] != "A":
                    return False
            return True
        else:
            return False
    except TypeError as e:
        return f"Value must be a string not {string}"
    except Exception as e:
        return e
    
    
print("A Program to validate whether a string is a valid\
Rondo Form")
print()

string_a, string_b, string_c = "ABACADAEAFAGAHAIAJA", "ABA", "ABBACCA"
string_d, string_e = "ACAC", "A"

print(f"{string_a} is a valid rondo form = {valid_rondo(string_a)}")
print(f"{string_b} is a valid rondo form = {valid_rondo(string_b)}")
print(f"{string_c} is a valid rondo form = {valid_rondo(string_c)}")
print(f"{string_d} is a valid rondo form = {valid_rondo(string_d)}")
print(f"{string_e} is a valid rondo form = {valid_rondo(string_e)}")
```

```python
"""Question 2:
Create a function that returns the whole of the first 
sentence which contains a specific word. Include the full
stop at the end of the sentence.

Examples

txt = "I have a cat. I have a mat. Things are going swell."

sentence_searcher(txt, "have") ➞ "I have a cat."

sentence_searcher(txt, "MAT") ➞ "I have a mat."

sentence_searcher(txt, "things") ➞ "Things are going swell."

sentence_searcher(txt, "flat") ➞ ""
"""

def sentence_searcher(sentence, word):
    """
    
    Returns the first sentence which contains the word
    given
    
    """
    try:
        split_sentence = sentence.split(".")
        for sub_sentence in range(len(split_sentence)):
            if word.lower() in split_sentence[sub_sentence] or word.title() in split_sentence[sub_sentence]:
                return split_sentence[sub_sentence].strip() + "."
        return ""
            
    except AttributeError as e:
        return "Invalid value present! Values must be str!"
    except Exception as e:
        return "values must be string", e
        
print("A Program to extract a sentence if the word is contained in the sentence")
print()

word_1, word_2, word_3, word_4 = "have", "MAT", "things", "flat"
sentence = "I have a cat. I have a mat. Things are going swell."

print(f"Sentence of the {word_1} in {sentence} = {sentence_searcher(sentence, word_1)}")
print(f"Sentence of the {word_2} in {sentence} = {sentence_searcher(sentence, word_2)}")
print(f"Sentence of the {word_3} in {sentence} = {sentence_searcher(sentence, word_3)}")
print(f"Sentence of the {word_4} in {sentence} = {sentence_searcher(sentence, word_4)}")

```

```python
"""Question 3:
Given a number, find the "round "of each digit of the 
number. An integer is called "round" if all its digits 
except the leftmost (most significant) are equal to zero.

    - Round numbers: 4000, 1, 9, 800, 90
    - Not round numbers: 110, 707, 222, 1001

Create a function that takes a number and returns the 
"round" of each digit (except if the digit is zero) as a 
string. Check out the following examples for more 
clarification.

Examples

sum_round(101) ➞ "1 100"

sum_round(1234) ➞ "4 30 200 1000"

sum_round(54210) ➞ "10 200 4000 50000"
"""

def sum_round(number):
    """
    
    Returns the "round" of each digit (except if the digit
    is zero) as a string.
    
    Examples

    sum_round(101) ➞ "1 100"

    sum_round(1234) ➞ "4 30 200 1000"

    sum_round(54210) ➞ "10 200 4000 50000"
    
    """
    if type(number) == int or type(number) == str:
        str_number = str(number)
        values = []
        zero = ""
        for digit in range(len(str_number)-1, -1, -1):
            if str_number[digit] != "0":
                values.append(str_number[digit] + zero)
                zero += "0"
            else:
                zero += "0"

        return " ".join(values)
    


print("A Program to display the round of each digit (except if the digit is zero) as a string.")
print()

num1, num2, num3 = 101, 1234, 54210
print(f"Round of {num1} = {sum_round(num1)}")
print(f"Round of {num2} = {sum_round(num2)}")
print(f"Round of {num3} = {sum_round(num3)}")
```

```python
"""Question 4:
Your task, is to create N x N multiplication table, of 
size n provided in parameter.

For example, when n is 5, the multiplication table is:

   - 1, 2, 3, 4, 5
   - 2, 4, 6, 8, 10
   - 3, 6, 9, 12, 15
   - 4, 8, 12, 16, 20
   - 5, 10, 15, 20, 25

This example will result in:

[[1, 2, 3, 4, 5], [2, 4, 6, 8, 10], [3, 6, 9, 12, 15], [4, 8, 12, 16, 20], [5, 10, 15, 20, 25]]

Examples

multiplication_table(1) ➞ [[1]]

multiplication_table(3) ➞ [[1, 2, 3], [2, 4, 6], [3, 6, 9]]
"""

def multiplication_table(number):
    """
    
    Returns a list of lists showing the multiplication
    table of a number.
    
    Examples

    multiplication_table(1) ➞ [[1]]

    multiplication_table(3) ➞ [[1, 2, 3], [2, 4, 6], [3, 6, 9]]
    
    """
    try:
        mul_table = []
        for value in range(1, number+1):
            row = []
            for digit in range(1, number+1):
                row.append(digit*value)
            mul_table.append(row)

        return mul_table
    except TypeError as e:
        return "Invalid value present! Values must be int!", e
    except Exception as e:
        return e

    
print("A Program to display the multiplication table of a number")
print()

num1, num2 = 1, 3
print(f"Multiplication table of {num1} = {multiplication_table(num1)}")
print(f"Multiplication table of {num2} = {multiplication_table(num2)}")
```

```python
"""Question 5:
Create a function that returns True if two lines rhyme 
and False otherwise. For the purposes of this exercise,
two lines rhyme if the last word from each sentence 
contains the same vowels.

Examples

does_rhyme("Sam I am!", "Green eggs and ham.") ➞ True

does_rhyme("Sam I am!", "Green eggs and HAM.") ➞ True
# Capitalization and punctuation should not matter.

does_rhyme("You are off to the races", "a splendid day.") ➞ False

does_rhyme("and frequently do?", "you gotta move.") ➞ False
"""

def does_rhyme(sentence_a, sentence_b):
    """
    
    Returns True if the last word of each sentence 
    contain the same vowels.
    
    Examples

    does_rhyme("Sam I am!", "Green eggs and ham.") ➞ True

    does_rhyme("Sam I am!", "Green eggs and HAM.") ➞ True
    # Capitalization and punctuation should not matter.

    does_rhyme("You are off to the races", "a splendid day.") ➞ False
    
    """
    try:
        if not(sentence_a[-1].isalpha()) and not(sentence_b[-1].isalpha()):
            if sentence_a[-2].lower() == sentence_b[-2].lower():
                return True
            return False

        elif sentence_a[-1].isalpha() and not(sentence_b[-1].isalpha()):
            if sentence_a[-1].lower() == sentence_b[-2].lower():
                return True
            return False

        elif not(sentence_a[-1].isalpha()) and sentence_b[-1].isalpha():
            if sentence_a[-2].lower() == sentence_b[-1].lower():
                return True
            return False

        else:
            return sentence_a[-1] == sentence_b[-1]
    except TypeError as e:
        return "Invalid value present! Values must be string!", e
    except Exception as e:
        return e


print("A Program to Check if two sentences rhyme")
print()

sent_a1, sent_a2 = "Sam I am!", "Green eggs and ham."
sent_b1, sent_b2 = "Sam I am!", "Green eggs and HAM."
sent_c1, sent_c2 = "You are off to the races", "a splendid day."
sent_d1, sent_d2 = "and frequently do?", "you gotta move."

print(f"'{sent_a1}' and '{sent_a2}' rhyme = {does_rhyme(sent_a1, sent_a2)}")
print(f"'{sent_b1}' and '{sent_b2}' rhyme = {does_rhyme(sent_b1, sent_b2)}")
print(f"'{sent_c1}' and '{sent_c2}' rhyme = {does_rhyme(sent_c1, sent_c2)}")
print(f"'{sent_d1}' and '{sent_d2}' rhyme = {does_rhyme(sent_d1, sent_d2)}")
```
