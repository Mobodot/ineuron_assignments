```python
"""Question 1:
Write a function that takes a positive integer num and 
calculates how many dots exist in a pentagonal shape 
around the center dot on the Nth iteration.

In the image below you can see the first iteration is 
only a single dot. On the second, there are 6 dots. On 
the third, there are 16 dots, and on the fourth there 
are 31 dots.








Return the number of dots that exist in the whole 
pentagon on the Nth iteration.

Examples

pentagonal(1) ➞ 1

pentagonal(2) ➞ 6

pentagonal(3) ➞ 16

pentagonal(8) ➞ 141
"""

def pentagonal(number):
    """
    
    Returns the number of dots that exist in a whole
    pentagon on the Nth iteration.
    
    E.g:
    pentagonal(1) ➞ 1
    pentagonal(2) ➞ 6
    pentagonal(3) ➞ 16
    pentagonal(8) ➞ 141
    
    """
    if type(number) == int:
        if number == 0:
            return -1
        elif number == 1:
            return 1
        else:
            return pentagonal(number-1) + 5*(number-1)
    return "Invalid input!", number
    
    
print("A Program to Find the Number of dots that exist in a whole Pentagon on Nth iteration.")

while True:
    try:
        print()
        number = int(input("Enter Nth iteration: "))
    except Exception as e:
        print("Invalid input! ")
    else:
        print(f"Number of dots in Pentagon on {number}th iteration = {pentagonal(number)}")
        break
```

```python
"""Question 2:
Make a function that encrypts a given input with these 
steps:

Input: "apple"

Step 1: Reverse the input: "elppa"
Step 2: Replace all vowels using the following chart:

a => 0
e => 1
i => 2
o => 2
u => 3

# "1lpp0"

Step 3: Add "aca" to the end of the word: "1lpp0aca"

Output: "1lpp0aca"

Examples

encrypt("banana") ➞ "0n0n0baca"

encrypt("karaca") ➞ "0c0r0kaca"

encrypt("burak") ➞ "k0r3baca"

encrypt("alpaca") ➞ "0c0pl0aca"

"""

def encrypt(string):
    """
    
    Returns a string which has been encrypted, by reversing
    it, replacing the vowels with a=0, e=1, i=2, o=2, u=3
    and adding aca at the end of the string.
    
    E.g
    encrypt("banana") ➞ "0n0n0baca"
    encrypt("karaca") ➞ "0c0r0kaca"
    encrypt("burak") ➞ "k0r3baca"
    encrypt("alpaca") ➞ "0c0pl0aca"
    
    """
    vowels = {
        "a": "0", "e": "1", "i": "2", "o": "2", "u": "4"
    }
    
    new_string = string[::-1]     # reversing the string
    for letter in new_string:
        if letter in vowels.values():     # checking if the vowel has already been replaced 
            continue
        elif letter in vowels:
            new_string = new_string.replace(letter, vowels[letter])
    return new_string + "aca"


print("A Program to Encrypt a String.")
print()

user_input = input("Enter a string: ")

print()
print(f"Encrypted version of {user_input} = {encrypt(user_input)}")
```

```python
"""Question 3:
Given the month and year as numbers, return whether that 
month contains a Friday 13th.
(i.e You can check Python's datetime module)

Examples

has_friday_13(3, 2020) ➞ True

has_friday_13(10, 2017) ➞ True

has_friday_13(1, 1985) ➞ False
"""

def has_friday_13(month, year):
    """
    
    Returns True if there is Fri 13th in a given month and 
    year, Otherwise return False
    
    """
    from datetime import date
    d = date(year=year, month=month, day=13)
    if d.isoweekday() == 5:     # Checks value of Weekdays where Monday is 1 and Sunday is 7
        return True
    return False


print("A Program to Check if 13th is Friday based on the given Month and Year.")

while True:
    try:
        print()
        month = int(input("Enter a month(1-12): "))
        year = int(input("Enter year(numeric): "))
        print(f"13th of month {month} in year {year} is Friday? = {has_friday_13(month, year)}")
        break
    except Exception as e:
        print("Invalid input! ", e)
```

```python
"""Question 4:
Write a regular expression that will help us count how 
many bad cookies are produced every day. You must use 
RegEx negative lookbehind.

Example

lst = ["bad cookie", "good cookie", "bad cookie", "good cookie", "good cookie"]
pattern = "yourregularexpressionhere"

len(re.findall(pattern, ", ".join(lst))) ➞ 2
"""

def count_bad_cookie(lst):
    """
    
    Retruns the count of bad cookies in a list.
    
    """
    import re
    pattern = "(?<!good) cookie,"     # regex negative lookbehind pattern
    return len(re.findall(pattern, ", ".join(lst)))


lst = ["bad cookie", "good cookie", "bad cookie", "good cookie", "good cookie"]
print("Count of bad cookies = ", count_bad_cookie(lst))
```

```python
"""Question 5:
Given a list of words in the singular form, return a set
of those words in the plural form if they appear more 
than once in the list.

Examples

pluralize(["cow", "pig", "cow", "cow"]) ➞ { "cows", "pig" }

pluralize(["table", "table", "table"]) ➞ { "tables" }

pluralize(["chair", "pencil", "arm"]) ➞ { "chair", "pencil", "arm" }
"""

def pluralize(lst):
    """
    
    Returns a set which contains the plural of a word if
    it appears more than once in a list otherwise, contains
    the singular version of the word.
    
    """
    plural = set()
    for value in lst:
        if lst.count(value) == 1:
            plural.add(value)
        else:
            plural.add(value+"s")
    return plural


lst1 = ["cow", "pig", "cow", "cow"]
lst2 = ["table", "table", "table"]
lst3 = ["chair", "pencil", "arm"]  
print(pluralize(lst1))
print(pluralize(lst2))
print(pluralize(lst3))
```
