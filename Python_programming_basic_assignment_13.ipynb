```python
"""Question 1: Write a program that calculates and 
prints the value according to the given formula:
Q = Square root of [(2 * C * D)/H]
Following are the fixed values of C and H:
C is 50. H is 30.
D is the variable whose values should be input to your 
program in a comma-separated sequence.

Example
Let us assume the following comma separated input 
sequence is given to the program: 100,150,180
The output of the program should be:
18,22,24
"""

def cdh_calculator(*args):
    """
    
    *args: Takes numeric values.
    
    This is a generator function which yields the
    calculation of numeric values using the formula:
    
    Q = Square root of [(2 * C * D)/H] 
    
    if a non-numeric value is given it raises an Exception.
    
    """
    import math
    
    c, h = 50, 30
    try:
        for value in args:
            if type(value) == list or type(value) == tuple:
                for number in value:
                    yield round(
                        math.sqrt((2 * c * number)/ h), 0)
            elif type(value) == int:
                yield round(
                        math.sqrt((2 * c * value)/ h), 0)
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
        

    
print("CDH Calculator. (Formular: Square root of [(2 * C * D)/H]), where C=30, H=50)")
print()
  
    
data = 100,150,180, [100,140,"180"]
print("Result of Calculation: ")
for value in cdh_calculator(data):
    print(value, end=", ")
    
```

```python
"""Question 2:
Write a program which takes 2 digits, X,Y as input and 
generates a 2-dimensional array. The element value in 
the i-th row and j-th column of the array should be i*j.

Note: i=0,1.., X-1; j=0,1,Â¡Y-1.

Example
Suppose the following inputs are given to the program:
3,5
Then, the output of the program should be:
[[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]] 
"""

def generate_data(row, col):
    """
    
    This function generates all the data for the logic 
    to be added to the matrix.
    
    i.e: The element value in the i-th row and j-th 
    column of the array should be i*j.
    
    e.g: generate_data(3,5) = 
    0,0,0,0,0,0, 0,1,2,3,4, 0,2,4,6,8
    
    """
    row_index = 0
    col_index = 0
    while True:
        if col_index < col:       # generate the values for cols      
            yield row_index * col_index
            col_index += 1
        elif col_index == col:    # creates data for new row
            row_index += 1
            col_index = 0
            if row_index == row:  # if all rows have been created
                break
            
def matrix(x, y):
    """
    
    x: rows
    y: cols
    
    Returns an (x, y) matrix of data
    
    """
    rows = []
    for value in range(x):
        rows.append([])
    
    count = 0
    row_index = 0
    for data in generate_data(x,y):
        if count < y:    # add col data to each row
            rows[row_index].append(data)
            count += 1
        else:
            row_index += 1    # moves to next row and add col data
            rows[row_index].append(data)
            count = 1
            
    return rows
            
row,col = 5,4
matrix(row,col)
```

```python
"""Question 3:
Write a program that accepts a comma separated sequence 
of words as input and prints the words in a 
comma-separated sequence after sorting them 
alphabetically.

Suppose the following input is supplied to the program:
without,hello,bag,world. Then, the output should be:
bag,hello,without,world
"""

def word_sort(lst_values):
    """
    
    Returns a sorted string of words available in a list
    
    """
    try:
        result = sorted(lst_values)
    except TypeError as e:
        print("Cannot perform sort operation with values of different types")
    except Exception as e:
        print(e)
    return ",".join(result)


print("Word Sorting Program")
print()

user_input = input("Enter words seperated by a comma: ")
split_input = user_input.split(",")

# removing spaces in words
stripped_split_input = []
for value in split_input:
    stripped_split_input.append(value.strip())
    
print(f"Sorted Result = {word_sort(stripped_split_input)}")
```

```python
"""Question 4: 
Write a program that accepts a sequence of whitespace 
separated words as input and prints the words after 
removing all duplicate words and sorting them 
alphanumerically.

Suppose the following input is supplied to the program:
hello world and practice makes perfect and hello world 
again.

Then, the output should be:
again and hello makes perfect practice world"""

def distinct_word_sort(lst_values):
    """
    
    Returns a sorted string of all the distinct words in
    a list.
    
    """
    distinct_words = set(lst_values)
    result = sorted(distinct_words)
    return " ".join(result)

print("Distinct Word Sorting Program")
print()

user_input = input("Enter words seperated by spaces: ")
split_input = user_input.split(" ")

# removing spaces in words
stripped_split_input = []
for value in split_input:
    stripped_split_input.append(value.strip())
    
print()
print(f"Sorted Result = {distinct_word_sort(stripped_split_input)}")
```

```python
"""Question 5: 
Write a program that accepts a sentence and calculate 
the number of letters and digits.

Suppose the following input is supplied to the program:
hello world! 123

Then, the output should be:
LETTERS 10
DIGITS 3
"""

def num_letters_and_digits(string):
    """
    
    Returns a tuple containing the number of letters,
    digits, and extras(neither letters or digits).
    
    """
    letters = 0
    digits = 0
    extras = 0
    for letter in string:
        if letter.isalpha():
            letters += 1
        elif letter.isnumeric():
            digits += 1
        else:
            extras += 1
    return letters, digits, extras


print("Letters and Digits Calculation Program")
print()

user_input = input("Enter words seperated by spaces: ")
    
print()
print(f"Number of LETTERS = {num_letters_and_digits(user_input)[0]}\n",
      f"Number of DIGITS = {num_letters_and_digits(user_input)[1]}\n",
     f"Number of EXTRAS = {num_letters_and_digits(user_input)[2]}")
```

```python
"""Question 6: 
A website requires the users to input username and 
password to register. Write a program to check the 
validity of password input by users.

Following are the criteria for checking the password:
1. At least 1 letter between [a-z]
2. At least 1 number between [0-9]
3. At least 1 letter between [A-Z]
4. At least 1 character from [$#@]
5. Minimum length of transaction password: 6
6. Maximum length of transaction password: 12

Your program should accept a sequence of comma separated 
passwords and will check them according to the above 
criteria. Passwords that match the criteria are to be 
printed, each separated by a comma.

Example:
If the following passwords are given as input to the 
program: ABd1234@1,a F1#,2w3E*,2We3345

Then, the output of the program should be:
ABd1234@1
"""

def valid_password(password_list):
    """
    
    Returns a list of passwords that meet the criteria 
    below:
    
    1. At least 1 letter between [a-z]
    2. At least 1 number between [0-9]
    3. At least 1 letter between [A-Z]
    4. At least 1 character from [$#@]
    5. Minimum length of transaction password: 6
    6. Maximum length of transaction password: 12
    
    """
    import re
    
    pattern = re.compile("^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[@$#])[A-Za-z0-9@$#]{6,12}$")
    
    passwords = []
    for password in password_list:
        if pattern.search(password):
            passwords.append(password)
    return passwords
    
    
print("Valid Passwords Program")
print()

user_input = input("Enter passwords seperated by spaces: ")
split_input = user_input.split(",")

# removing spaces in words
stripped_split_input = []
for password in split_input:
    stripped_split_input.append(password.strip())

print()
print(f"List of correct passwords = {validate_password(stripped_split_input)}")
```
