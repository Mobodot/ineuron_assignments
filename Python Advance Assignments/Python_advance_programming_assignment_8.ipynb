```python
"""Question 1:
Given a sentence as txt, return True if any two adjacent 
words have this property: One word ends with a vowel, 
while the word immediately after begins with a vowel 
(a e i o u).

Examples

vowel_links("a very large appliance") ➞ True

vowel_links("go to edabit") ➞ True

vowel_links("an open fire") ➞ False

vowel_links("a sudden applause") ➞ Fals
"""

def vowel_links(sentence):
    """
    
    Returns True if the ending and starting of any two 
    adjacent words are vowels
    
    Examples

    vowel_links("a very large appliance") ➞ True

    vowel_links("go to edabit") ➞ True

    vowel_links("an open fire") ➞ False
    
    """
    vowels = ("a", "e", "i", "o", "u")
    for index in range(len(sentence)):
        if sentence[index].isspace():     # check if the letter before and after the space are vowels
            if (sentence[index-1] in vowels) and (sentence[index+1] in vowels):
                return True
            
    return False


print("A Program to Return True One word ends with a vowel, while the word immediately after begins with a vowel  ")
print()

user_input = input("Enter a string: ")
print()
print(f"\"{user_input}\" contains vowel links = {vowel_links(user_input)}")
```

```python
"""Question 2: 
You are given three inputs: a string, one letter, and a 
second letter.

Write a function that returns True if every instance of
the first letter occurs before every instance of the 
second letter.

Examples

first_before_second("a rabbit jumps joyfully", "a", "j") ➞ True
# Every instance of "a" occurs before every instance of "j".

first_before_second("knaves knew about waterfalls", "k", "w") ➞  True

first_before_second("happy birthday", "a", "y") ➞ False
# The "a" in "birthday" occurs after the "y" in "happy".

first_before_second("precarious kangaroos", "k", "a") ➞ False
"""

def first_before_second(string, f_letter, s_letter):
    """
    
    f_letter: str
    s_letter: str
    
    Returns True if every instance of the first letter
    occurs before every instance of the second letter.
    Otherwise False
    
    Examples

    first_before_second("a rabbit jumps joyfully", "a", "j") ➞ True
    # Every instance of "a" occurs before every instance of "j".
    
    first_before_second("happy birthday", "a", "y") ➞ False
    # The "a" in "birthday" occurs after the "y" in "happy".

    
    """
    s_letter_index = string.index(s_letter)
    for index in range(len(string)):
        if string[index] == f_letter:
            if index > s_letter_index:
                return False
    return True
            

print(first_before_second("a rabbit jumps joyfully", "a", "j"))
print(first_before_second("knaves knew about waterfalls", "k", "w"))
print(first_before_second("happy birthday", "a", "y"))
print(first_before_second("precarious kangaroos", "k", "a"))
```

```python
"""Question 3: 
Create a function that returns the characters from a list
or string r on odd or even positions, depending on the 
specifier s. The specifier will be "odd" for items on odd
positions (1, 3, 5, ...) and "even" for items on even 
positions (2, 4, 6, ...).

Examples

char_at_pos([2, 4, 6, 8, 10], "even") ➞ [4, 8]
# 4 & 8 occupy the 2nd & 4th positions

char_at_pos("EDABIT", "odd") ➞ "EAI"
# "E", "A" and "I" occupy the 1st, 3rd and 5th positions

char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd") ➞ ["A", "B", "T", "A", "I", "Y"]
"""
    
def char_at_pos(sequence, string):
    """
    
    string: odd or even
    
    Returns the characters from a list or string on odd or
    even positions, depending on the specifier string
    
    """
    if type(sequence) == str or type(sequence) == list:
        if string.lower() == "odd":
            values = [sequence[index-1] for index in range(1, len(sequence)+1) if index%2 != 0]
        elif string.lower() == "even":
            values = [sequence[index-1] for index in range(1, len(sequence)+1) if index%2 == 0]
        else:
            return "string value must be either 'odd' or 'even'"
    
    
        if type(sequence) == list:
            return values
        else:
            return "".join(values)
    else:
        return "Invalid sequence!"
    
    
print(char_at_pos([2, 4, 6, 8, 10], "even"))
print(char_at_pos("EDABIT", "odd"))
print(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd")
```

```python
"""Question 4:
Write a function that returns the greatest common divisor
of all list elements. If the greatest common divisor is 
1, return 1.

Examples

GCD([10, 20, 40]) ➞ 10

GCD([1, 2, 3, 100]) ➞ 1

GCD([1024, 192, 2048, 512]) ➞ 64
"""

def modulus(dividend, divisor):
    """
    
    Returns True if a modulus of the dividend and divisor
    is 0 otherwise False
    
    """
    if dividend % divisor == 0:
        return True
    return False

    
def GCD(lst):
    """
    
    Returns the Greatest common divisor of all list elements
    
    """
    if type(lst) == list:
        smallest = min(lst)
        for value in range(1, smallest+1):
            counter = 0
            for number in range(len(lst)):
                if modulus(lst[number], value) == True:
                    counter += 1
            if counter == len(lst):
                hcf = value
        return hcf
    else:
        return "Invalid input"
    
    
print(GCD([10, 20, 40]))
print(GCD([1, 2, 3, 100]))
print(GCD([1024, 192, 2048, 512]))
```

```python
"""Question 5:
A number/string is a palindrome if the digits/characters 
are the same when read both forward and backward. Examples 
include "racecar" and 12321. Given a positive number n, 
check if n or the binary representation of n is 
palindromic. Return the following:

- "Decimal only." if only n is a palindrome.
- "Binary only." if only the binary representation of n 
is a palindrome.
- "Decimal and binary." if both are palindromes.
- "Neither!" if neither are palindromes.

Examples

palindrome_type(1306031) ➞ "Decimal only."
# decimal = 1306031
# binary  = "100111110110110101111"

palindrome_type(427787) ➞ "Binary only."
# decimal = 427787
# binary  = "1101000011100001011"

palindrome_type(313) ➞ "Decimal and binary."
# decimal = 313
# binary  = 100111001

palindrome_type(934) ➞ "Neither!"
# decimal = 934
# binary  = "1110100110"
"""

def palindrome_type(number):
    """
    
    Return "Decimal and binary" if the binary and decimal
    representation of a number are the same when read both
    forward and backward. Otherwise the only the binary
    representation is same return "Binary only", if only the
    Decimal representation is the same return "Decimal Only".
    If both decimal and binary representation are not the 
    same return "Neither!"
    
    Examples

    palindrome_type(1306031) ➞ "Decimal only."
    # decimal = 1306031
    # binary  = "100111110110110101111"

    palindrome_type(427787) ➞ "Binary only."
    # decimal = 427787
    # binary  = "1101000011100001011"

    palindrome_type(313) ➞ "Decimal and binary."
    # decimal = 313
    # binary  = 100111001

    palindrome_type(934) ➞ "Neither!"
    # decimal = 934
    # binary  = "1110100110"
    
    """
    str_number = str(number)
    binary = bin(number)[2:]
    
    if str_number == str_number[::-1] and binary == binary[::-1]:
        return "Decimal and binary"
    elif str_number == str_number[::-1]:
        return "Decimal only"
    elif binary == binary[::-1]:
        return "Binary only"
    else:
        return "Neither!"
        
    

print(palindrome_type(1306031))
print(palindrome_type(427787))
print(palindrome_type(313))
print(palindrome_type(934))
```
