```python
"""Question 1:
For this challenge, forget how to add two numbers 
together. The best explanation on what to do for this 
function is this meme:


Examples

meme_sum(26, 39) ➞ 515
# 2+3 = 5, 6+9 = 15
# 26 + 39 = 515

meme_sum(122, 81) ➞ 1103
# 1+0 = 1, 2+8 = 10, 2+1 = 3
# 122 + 81 = 1103

meme_sum(1222, 30277) ➞ 31499
"""

def meme_sum(num1, num2):
    """
    
    Returns the meme_sum of two numbers.
    
    Examples

    meme_sum(26, 39) ➞ 515
    # 2+3 = 5, 6+9 = 15
    # 26 + 39 = 515

    meme_sum(122, 81) ➞ 1103
    # 1+0 = 1, 2+8 = 10, 2+1 = 3
    # 122 + 81 = 1103
    
    """
    str_num1 = str(num1)
    str_num2 = str(num2)
    meme_total = []
    
    # if the 2 numbers are not of the same len add 0 at the begining of the lesser number
    if len(str_num1) > len(str_num2):
        diff = len(str_num1) - len(str_num2)
        str_num2 = "0"*diff + str_num2
    elif len(str_num2) > len(str_num1):
        diff = len(str_num2) - len(str_num1)
        str_num1 = "0"*diff + str_num1
    else:
        pass
    
    try:
        # meme addition
        for digit in range(len(str_num1)):
            meme_total.append(str(int(str_num1[digit]) + int(str_num2[digit])))
        return int("".join(meme_total))
    except TypeError as e:
        return "Values must be integers only!"
    except Exception as e:
        return e
    
    
print("A Program to Calculate the Meme Sum of Two Numbers")
print()

num1a, num1b = 26, 39
num2a, num2b = 122, 81
num3a, num3b = 1222, 30277
print(f"Meme sum of {num1a, num1b} = {meme_sum(num1a, num1b)}")
print(f"Meme sum of {num2a, num2b} = {meme_sum(num2a, num2b)}")
print(f"Meme sum of {num3a, num3b} = {meme_sum(num3a, num3b)}")
```

```python
"""Question 4:
The Kempner Function, applied to a composite number, 
permits to find the smallest integer greater than zero 
whose factorial is exactly divided by the number.

kempner(6) ➞ 3

1! = 1 % 6 > 0
2! = 2 % 6 > 0
3! = 6 % 6 === 0

kempner(10) ➞ 5

1! = 1 % 10 > 0
2! = 2 % 10 > 0
3! = 6 % 10 > 0
4! = 24 % 10 > 0
5! = 120 % 10 === 0

A Kempner Function applied to a prime will always 
return the prime itself.

kempner(2) ➞ 2
kempner(5) ➞ 5

Given an integer n, implement a Kempner Function.

Examples

kempner(6) ➞ 3

kempner(10) ➞ 5

kempner(2) ➞ 2
"""

def factorial(number):
    """
    
    Returns the factorial of a number
    
    """
    if number <= 1:
        return 1
    return number * factorial(number-1)


def is_prime(number):
    """
    
    Returns True if a number is a prime number otherwise
    False.
    
    """
    if number <= 1:
        return False
    for digit in range(2, number):
        if number % digit == 0:
            return False
    return True


def kempner(number):
    """
    
    Returns the smallest integer greater than zero 
    whose factorial is exactly divided by the number. 
    
    kempner(6) ➞ 3

    1! = 1 % 6 > 0
    2! = 2 % 6 > 0
    3! = 6 % 6 === 0

    kempner(10) ➞ 5

    1! = 1 % 10 > 0
    2! = 2 % 10 > 0
    3! = 6 % 10 > 0
    4! = 24 % 10 > 0
    5! = 120 % 10 === 0
    
    If the number is a prime number it returns that number.
    
    Examples

    kempner(6) ➞ 3

    kempner(10) ➞ 5

    kempner(2) ➞ 2
    
    kempner(5) ➞ 5
    
    
    """
    try:
        if is_prime(number):
            return number
        else:
            for value in range(number):
                if factorial(value) % number == 0:
                    return value
    except TypeError as e:
        return f"Value must be int not {type(number)}"
    except Exception as e:
        return e
    
print("----- A Program to Output the Kempner Number of a Number----")
print()

num1, num2, num3, num4 = 6, 10, 5, 2
print(f"Kempner Number of {num1} = {kempner(num1)}")
print(f"Kempner Number of {num2} = {kempner(num2)}")
print(f"Kempner Number of {num3} = {kempner(num3)}")
print(f"Kempner Number of {num4} = {kempner(num4)}")
```
