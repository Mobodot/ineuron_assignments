```python
"""Question 1 :
Write a function that takes a list of lists and returns
the value of all of the symbols in it, where each symbol
adds or takes something from the total score. Symbol 
values:

# = 5
O = 3
X = 1
! = -1
!! = -3
!!! = -5

A list of lists containing 2 #s, a O, and a !!! would 
equal (0 + 5 + 5 + 3 - 5) 8.

If the final score is negative, return 0 
(e.g. 3 #s, 3 !!s, 2 !!!s and a X would be 
(0 + 5 + 5 + 5 - 3 - 3 - 3 - 5 - 5 + 1) -3, so return 0.

Examples

check_score([
  ["#", "!"],
  ["!!", "X"]
]) ➞ 2

check_score([
  ["!!!", "O", "!"],
  ["X", "#", "!!!"],
  ["!!", "X", "O"]
]) ➞ 0
"""

def check_score(lst_of_lst):
    """
    
    Returns an int value which is the evaluation of symbols 
    in a list of of lists.
    If result is negative returns 0.
    
    """
    symbol_values = {
        "#": 5,
        "O": 3,
        "X": 1,
        "!": -1,
        "!!": -3,
        "!!!": -5
    }
    
    total = 0
    for lst in lst_of_lst:
        if type(lst) == list:     # checks if the list contains only lists
            for symbol in lst:
                if symbol in symbol_values:     # checks if symbol is correct
                    total += symbol_values[symbol]
                else:
                    return "Invalid Symbol Present!"
        else:
            return "Value is not of type list"
            
    if total < 1:
        return 0
    else:
        return total


lst1 = [["!!!", "O", "!"], ["X", "#", "!!!"],["!!", "X", "O"]]
lst2 = [["#", "!"], ["!!", "X"]]

print("Result for lst1 = ", check_score(lst1))
print("Result for lst2 = ", check_score(lst2))


```

```python
"""Question 2:
Create a function that takes a variable number of 
arguments, each argument representing the number of items
in a group, and returns the number of permutations 
(combinations) of items that you could get by taking one 
item from each group.

Examples

combinations(2, 3) ➞ 6

combinations(3, 7, 4) ➞ 84

combinations(2, 3, 4, 5) ➞ 120
"""

def combinations(*args):
    """
    
    args: int values
    
    Returns the the number of permutations based on the
    given args.
    
    """
    total = 1
    for value in args:
        if type(value) == int:
            total *= value
        else:
            return "Values must be int only!"
    return total


combinations(3, 7, 4)
```

```python
"""Question 4:
Write a function that takes a number and returns True if 
it's a prime; False otherwise. The number can be 
2^64-1 (2 to the power of 63, not XOR). 
With the standard technique it would be O(2^64-1), which 
is much too large for the 10 second time limit.

Examples

prime(7) ➞ True

prime(56963) ➞ True

prime(5151512515524) ➞ False
"""

def prime(number):
    """
    
    Returns True if number is a Prime number otherwise,
    False
    
    """
    if type(number) == int or type(number) == float:
        if number==2 or number==3:
            return True
        if number%2==0 or number<2:     # 0 and 1 are not prime numbers
            return False
        for value in range(3, int(number**1/2)+1, 2):     # Even numbers can't be prime numbers so only check odd numbers 
            if number%value==0:
                return False    
        return True
    else:
        return "Invalid value entered!", number
    
num1 = "7"
num2 = 56963
num3 = 5151512515524

print(prime(num1))
print(prime(num2))
print(prime(num3))
```
